```python
predict = data.groupby("Disease")[[
    "BloodPressure",
    "GlucoseLevel",
    "Cholesterol",
    "HeartRate",
    "BMI"
]].mean()

predict.plot(kind="bar")
plt.title("Prediction of disease")
plt.ylabel("Disease")
plt.xticks(rotation=0)
plt.show()
```
